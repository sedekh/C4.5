```python
#cetak data ke table dengan DataFrame pandas
df = pd.read_csv('tenis.csv')
df
```

```python
#cetak info dataset
df.info()
```

```python
#encoder to int
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
#ubah semua tipe data keint
for column in df:
  if df[column].dtypes == object:
   df[column]=le.fit_transform(df[column])
```

```python
#cek tipe data
df.info()
```

```python
df
```

```python
#pisahkan kolom targe/label/kelas dalam hal ini adalah kolom main ke var x
#selebihnya ke var y
#x(:,:,-1) tanda : pertama dibaca kolom pertama, tanda : kolom terakir dikurang -1 (untuk kolom main)
x = df.iloc[:,:-1].values
y = df.iloc[:,-1].values
```

```python
#mengeluarkan kolom main pada tabel, karena nnti akan jadi target (var x)
#selebihnya var y
atr_df= df.drop(columns ='Main')
atr_df
```

```python
#mengeluarkan target yaitu kolom main
cls_df =df['Main']
cls_df
```

```python
#Model seleksi
#untuk split data train dan test
#cm
#algoritman c4.5
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
from sklearn.tree import DecisionTreeClassifier
```

```python
#split data
#random size = 0.2 berarti 20%
#random_state=1 kali putaran, boleh juga none atau 42
xtrain, xtest, ytrain, ytest = train_test_split(atr_df, cls_df, test_size=0.2, random_state=1)
tree_df = DecisionTreeClassifier(random_state=1)
tree_df.fit(xtrain, ytrain)
```

```python
#prediksi
y_pred = tree_df.predict(xtest)
cm = confusion_matrix(ytest, y_pred)
print("Confusion Matrix")
print(cm)
akurasi = classification_report(ytest, y_pred)
print("Akurasi C4.5")
print("Akurasi :", akurasi)
akurasi = accuracy_score(ytest, y_pred)
print("Akurasi : %d persen" %(akurasi*100))
```

```python
#buat grafik tree
from sklearn.tree import export_graphviz
export_graphviz(tree_df, out_file="Tree_Tenis.dot", class_names=["Ya","Tidak"],
                feature_names=atr_df.columns, impurity=False, filled=True)
```

```python
#baca grap, ubah dari dot ke pdf
import graphviz

with open("Tree_Tenis.dot") as fig:
  dot_graph = fig.read()
  graph = graphviz.Source(dot_graph)

#ubah ke pdf
graph.view()
```
